## Extracci√≥n de datos

```python
import pandas as pd
import numpy as np

url_data = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
```

```python
df = pd.read_json(url_data)
df
```

```python
df.info()
```

## Transformaci√≥n de datos

### Normalizaci√≥n de la tabla y datos anidados

```python
customer_df = pd.json_normalize(df['customer'])
phone_df = pd.json_normalize(df['phone'])
internet_df = pd.json_normalize(df['internet'])
account_df = pd.json_normalize(df['account'])

df = pd.concat([df, customer_df, phone_df, internet_df, account_df], axis=1)

df.head()
```

```python
df = df.drop(columns=['customer', 'phone', 'internet', 'account'])
df = df.rename(columns={'Charges.Monthly': 'MonthlyCharges', 'Charges.Total': 'TotalCharges'})

df.head()
```

```python
df.info()

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

categorical_cols = [
    'Churn', 'gender', 'SeniorCitizen', 'Partner', 'Dependents',
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
    'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'
]

for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')

df.info()
```

### Busqueda de errores o datos faltantes en las columnas

```python
# Buscar valores faltantes
print("Datos faltantes por columna:")
print(df.isnull().sum())

# Analisis en columnas categoricas para buscar errores
print("\nAnalisis de columnas categoricas:")
for col in categorical_cols:
    if col in df.columns:
        print(f"\nColumn: {col}")
        print(df[col].value_counts())
```

### Tratamiento de datos faltantes en ['TotalCharges'] y ['Churn']

```python
df['TotalCharges'] = df.apply(
    lambda row: row['MonthlyCharges'] * row['tenure'] if pd.isna(row['TotalCharges']) else row['TotalCharges'],
    axis=1
)

print("Valores faltantes despues de la funci√≥n:")
print(df.isnull().sum())
```

```python
df['Churn'] = df['Churn'].replace('', np.nan)
df.dropna(subset=['Churn'], inplace=True)
df['Churn'] = df['Churn'].astype('category')

print("Valores √∫nicos en 'Churn' despues de la limpieza:", df['Churn'].unique())
print("\nConteo de valores en 'Churn' despues de la limpieza:")
display(df['Churn'].value_counts())
```

### Columna de cuentas diarias

#### Se a√±adio una columna llamada 'Cuentas_Diarias' basado en el gasto mensual con un estimado de 30 dias para el mes.

```python
df['Cuentas_Diarias'] = (df['MonthlyCharges'] / 30).round(2)
df[['MonthlyCharges', 'Cuentas_Diarias']].head()
```

```python
datos_tratados = df.copy()

for col in datos_tratados.columns:
    if datos_tratados[col].dtype == 'category':
        datos_tratados[col] = datos_tratados[col].cat.rename_categories({'Yes': 1, 'No': 0})
    elif datos_tratados[col].dtype == 'object':
        datos_tratados[col] = datos_tratados[col].replace({'Yes': 1, 'No': 0})

datos_tratados.head()
```

```python
datos_tratados.to_csv('datos_tratados.csv', index=False)
```
